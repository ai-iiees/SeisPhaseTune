```python
from obspy import read_inventory
```

```python
inv = read_inventory(r'c:\Users\ikahbasi\OneDrive\Applications\Seismic-Response\BI_2022-09-14.dataless')
```

```python
stations_passed = {""}
with open('IIEES-BIN.csv', 'w') as output:
    output.write('sensor,network,station,location,channel,latitude,longitude,elevation,region\n')
    for net in inv:
        for sta in net:
            if sta.code in stations_passed:
                continue
            loc = sta.channels[0].location_code
            cha = sta.channels[0].code[:2]
            txt = f'guralp,{net.code},{sta.code},{loc},{cha},{sta.latitude},{sta.longitude},{sta.elevation},IRAN\n'
            print(txt)
            output.write(txt)
            stations_passed.add(sta.code)
            # break
```
